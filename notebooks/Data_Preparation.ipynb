```python
import pandas as pd
import os
from shutil import move

dicom_data = pd.read_csv('csv/dicom_info.csv')
dicom_data.head()
```

```python
Calc_Training = pd.read_csv('csv/calc_case_description_train_set.csv')
Calc_Test = pd.read_csv('csv/calc_case_description_test_set.csv')
Mass_Training = pd.read_csv('csv/mass_case_description_train_set.csv')
Mass_Test = pd.read_csv('csv/mass_case_description_test_set.csv')
```

```python
seriesDescriptions = ['cropped images', 'full mammogram images','ROI mask images']
i = 1
for index, row in dicom_data.iterrows():
    image_path = row['image_path']
    seriesDescription = row['SeriesDescription']
    if seriesDescription in seriesDescriptions:
      dataType = row['PatientID'].split('_')[0]
      dataType = dataType.replace('-', '_')
      patient_id = row['PatientID'].split('_')[1] + '_' + row['PatientID'].split('_')[2]
      if dataType in globals():
        df = globals()[dataType]
        row = df[df['patient_id'] == patient_id]
        # Check if the row was found
        if not row.empty:
            # Get the corresponding value in the 'pathology' column
            corresponding_pathology = row['pathology'].iloc[0]
        new_path = f'Images/{seriesDescription}/{corresponding_pathology}'
        new_folder = os.path.join(new_path, f'{i}.jpg')
        os.makedirs(new_path, exist_ok=True)
        i += 1
        move(image_path, new_folder)
        print(f'{image_path} - {corresponding_pathology}')
```
